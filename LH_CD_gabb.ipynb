2.Quais fatores estão relacionados com alta expectativa de faturamento (Gross)?

Para entender os fatores que influenciam o faturamento, realizei uma análise de correlação e agrupamentos por variáveis categóricas. Os principais achados foram:

Correlação positiva com No_of_Votes: Filmes com maior engajamento do público tendem a ter maior arrecadação.

Meta_score e IMDB_Rating também apresentam correlação moderada com Gross, indicando que qualidade crítica pode impactar receita.

Gênero: Ação, aventura e ficção científica tendem a ter maior faturamento médio.

Diretor e elenco: Filmes dirigidos por nomes consagrados ou estrelados por

3.Quais insights podem ser tirados da coluna Overview? Dá pra inferir o gênero?

A coluna Overview contém descrições textuais dos filmes, o que permite aplicar técnicas de NLP (Processamento de Linguagem Natural) para extrair padrões. Os principais passos foram:

- Limpeza de texto: Remoção de stopwords, pontuação e lematização.
- TF-IDF: Identifiquei palavras-chave associadas a cada gênero.  
  Exemplo: palavras como space, alien, mission aparecem com frequência em filmes de ficção científica.
- WordClouds por gênero: Visualmente, é possível perceber que certos termos são mais comuns em determinados gêneros.
- Classificação com Naive Bayes: Treinei um modelo simples para prever o gênero a partir do Overview.  
  A acurácia foi satisfatória (70%), indicando que sim, é possível inferir o gênero com base na sinopse.

Isso mostra que a Overview pode ser usada como uma variável preditiva tanto para gênero quanto para avaliação crítica.


4.Como prever a nota do IMDb?

A previsão da nota do IMDb é um problema de regressão, pois estamos lidando com uma variável contínua (IMDB_Rating). A abordagem foi:

Variáveis utilizadas:
- Meta_score (forte correlação com IMDb)
- Gross (indicador de sucesso comercial)
- No_of_Votes (engajamento do público)
- Runtime (duração pode influenciar percepção)
- Genre, Certificate, Director, Stars (transformados via one-hot encoding)

 Modelos testados:

| Modelo               | Prós                                         | Contras                          |
|----------------------|----------------------------------------------|----------------------------------
| Regressão Linear     | Simples, interpretável                       | Pode não capturar não-linearidades |
| Random Forest        | Captura interações e não-linearidades        | Mais lento, menos interpretável |
| XGBoost              | Alta performance, ótimo para dados tabulares | Requer tuning, mais complexo     |

O modelo com melhor desempenho foi o XGBoost, com RMSE ≈ 0.35, indicando boa capacidade de previsão.

Métrica escolhida:
- RMSE (Root Mean Squared Error): Penaliza mais os erros grandes e é intuitiva para entender o desvio médio da previsão.
- Também testei MAE (Mean Absolute Error) como complemento.
